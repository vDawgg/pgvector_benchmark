```python
import matplotlib.pyplot as plt
from datasets import load_dataset, load_from_disk
```

```python
ds_v1 = load_from_disk('../data/dataset/train_dataset.json')
ds_v2 = load_dataset('microsoft/ms_marco', 'v2.1', split="train")
```

```python
ds_v1_passage_len = [len(p) for p in ds_v1["passages"]]
ds_v2_passage_len = [len(p["passage_text"]) for p in ds_v2["passages"]]
```
