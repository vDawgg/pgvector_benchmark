```python
import os
import pickle

from networkx.algorithms.bipartite.basic import color

results_dir = '../benchmark/results'

insert_log = pickle.load(open(os.path.join(results_dir, 'item_log.pkl'), 'rb'))
query_log = pickle.load(open(os.path.join(results_dir, 'query_log.pkl'), 'rb'))
```

```python
item_start, item_end = zip(*insert_log)
query_start, query_end, answers = zip(*query_log)
```

```python
import numpy as np

item_delta = np.subtract(np.array(item_end), np.array(item_start))
query_delta = np.subtract(np.array(query_end), np.array(query_start))
```

```python
import matplotlib.pyplot as plt

plt.plot(item_delta, color='red')
plt.plot(query_delta, color='blue')
plt.yscale('log')
plt.show()
```
